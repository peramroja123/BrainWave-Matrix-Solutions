#**EDA**
###**Univariate**
Univariate analysis examines one variable at a time to understand its distribution, central tendency, and variability.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for the visualizations
sns.set(style="whitegrid")

# Numerical Data Visualizations
numerical_cols = ['Sales', 'Quantity', 'Discount', 'Profit']

plt.figure(figsize=(10, 15))
for i, col in enumerate(numerical_cols):
    # Histogram
    plt.subplot(4, 2, i * 2 + 1)
    sns.histplot(df[col], kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
plt.tight_layout()
plt.show()
```

* Distribution of Sales: Most sales values are small, with very few orders having high sales amounts.

* Distribution of Quantity: Most orders involve small quantities (2-4 items), with larger quantities being rare.

* Distribution of Discount: Discounts are typically either small (close to 0%) or around 20%, with few orders receiving very high discounts.
* Distribution of Profit: Most orders generate small profits, while some incur significant losses or very high profits.

```python
# Categorical Data Visualizations
categorical_cols = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']

# Bar plots
plt.figure(figsize=(14, 10))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(3, 2, i)
    sns.countplot(y=df[col], order=df[col].value_counts().index, hue=df[col], palette='viridis',dodge=False, legend=False)
    plt.title(f'Frequency of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

* Ship Mode: Most orders are shipped using "Standard Class."
* Segment: The "Consumer" segment has the highest number of orders.
* Region: The "West" region has the most orders compared to other regions.
* Category: "Office Supplies" is the most frequently ordered category.
* Sub-Category: "Binders" and "Paper" are the top sub-categories by order count.

##**Bivariate Analysis**

Bivariate analysis examines the relationship between two variables to identify patterns, correlations

## Numeric vs Numeric

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Numeric columns
numeric_columns = ['Sales', 'Profit', 'Discount']

plt.figure(figsize=(16, 5))

# Iterate over pairs of numeric columns
pairs = [(numeric_columns[0], numeric_columns[1]),
         (numeric_columns[0], numeric_columns[2]),
         (numeric_columns[1], numeric_columns[2])]

for i, (x_col, y_col) in enumerate(pairs, start=1):
    plt.subplot(1, 3, i)
    binned_data = df.copy()
    binned_data['x_binned'] = pd.cut(df[x_col], bins=10, labels=False)
    sns.barplot(x='x_binned', y=y_col, data=binned_data, errorbar=None, alpha=0.8)
    plt.title(f"{x_col} vs {y_col}")
    plt.xlabel(f"Binned {x_col}")
    plt.ylabel(f"Mean {y_col}")

plt.tight_layout()
plt.show()
```

* Sales vs Profit: Higher sales generally lead to higher profits, except at the extremes where profit drops.
* Sales vs Discount: Higher sales tend to be associated with higher discounts.
* Profit vs Discount: Higher profits are associated with lower discounts.

```python
numerical_cols = ['Sales', 'Quantity', 'Discount', 'Profit']
plt.figure(figsize=(8, 6))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

* The heatmap shows that sales and profit have a moderate positive correlation (0.48), while discount and profit have a weak negative correlation (-0.22).

## Categorical vs Numerical

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Category', y='Profit', hue='Category', palette='Set2')
plt.title('Profit Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

* The graph shows that the Technology category generates significantly higher profits compared to Furniture and Office Supplies.

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Region', y='Sales', estimator='mean', errorbar=None, hue='Region', palette='muted')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

* The graph shows that sales are fairly consistent across all regions, with slight variations.

## Categorical vs Categorical

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the theme for seaborn plots
sns.set_theme(style="whitegrid")

# Ship Mode vs. Segment
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Ship Mode', hue='Segment')
plt.title('Distribution of Ship Mode across Segments')
plt.xlabel('Ship Mode')
plt.ylabel('Segment Count')
plt.legend(title='Segment')
plt.show()

# Category vs. Region
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Category', hue='Region')
plt.title('Distribution of Product Categories across Regions')
plt.xlabel('Category')
plt.ylabel('Region Count')
plt.legend(title='Region')
plt.show()

# Sub-Category vs. Segment
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Sub-Category', hue='Segment', order=df['Sub-Category'].value_counts().index)
plt.title('Distribution of Sub-Categories across Segments')
plt.xlabel('Sub-Category')
plt.ylabel('Segment Count')
plt.xticks(rotation=45)
plt.legend(title='Segment')
plt.show()
```

* The "Standard Class" shipping mode is the most commonly used across all segments, with "Consumer" being the largest contributor.
* Office Supplies are the most distributed category across all regions, with the West and East regions having the highest counts.
* "Binders" and "Paper" are the most purchased sub-categories across all segments, with "Consumer" leading in both.
